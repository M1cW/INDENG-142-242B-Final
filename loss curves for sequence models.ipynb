```python
import os
import pandas as pd
import matplotlib.pyplot as plt

history_path = os.path.join(
    "./simple_transformer_simple_return_t1_model",
    "training_history.csv"
)

history_df = pd.read_csv(history_path)

plt.figure(figsize=(10, 5))

plt.plot(
    history_df["epoch"],
    history_df["train_loss"],
    label="Training Loss",
    linewidth=2
)

plt.plot(
    history_df["epoch"],
    history_df["val_loss_raw_simple_return"],
    label="Validation Loss",
    linewidth=2
)

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss vs Validation Loss")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
